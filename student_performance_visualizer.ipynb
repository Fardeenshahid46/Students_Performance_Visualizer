```python
import pandas as pd
import matplotlib.pyplot as plt

#To Load Dataset
df=pd.read_csv("students.csv")

#To Calculate Average of marks
df['Average'] = df[['Math', 'English', 'Science']].mean(axis=1)

#Top Performers
top_students=df.sort_values(by="Average",ascending=False).head(5)

#Plot-Top 5 Students
plt.figure(figsize=(8,5))
plt.bar(top_students["Name"],top_students["Average"],color="purple")
plt.title("Top 5 Students by Average Score")
plt.ylabel("Average Marks")
plt.xlabel("Student Name")
plt.tight_layout()
plt.savefig("top_students.png")
plt.show()
```
